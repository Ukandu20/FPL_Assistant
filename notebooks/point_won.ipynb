```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import plotly as py
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

#import plotly.io as pio
#pio.renderers.default = "browser"

from dash import Dash, dcc, html, Input, Output

pd.set_option("display.max_columns", None)




#import csvs
pl_1920 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2019-2020/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2021 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2020-2021/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2122 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2021-2022/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2223 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2022-2023/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2324 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2023-2024/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2425 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2024-2025/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
pl_2526 = pd.read_csv("../data/processed/fbref/ENG-Premier League/2025-2026/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))

consolidate = pd.read_csv("../data/processed/fbref/ENG-Premier League/consolidated/team_match/schedule.csv", usecols=('league', 'season', 'team', 'game', 'date', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent','is_relegated'))
display(pl_1920.columns)

#team map to assign shorthand to opponents



# Traditional Big 6
big_6 = ['ARS', 'MUN', 'CHE', 'LIV', 'TOT', 'MCI']
relegated = pl_1920['is_relegated'] == 1

# Create a column indicating membership of the team in the traditional "big 6"
pl_1920['b6'] = pl_1920['team'].isin(big_6)
b6 = pl_1920['b6']

consolidate['b6'] = consolidate['team'].isin(big_6)
b6 = consolidate['b6']

display(pl_1920[relegated])
display(pl_1920[b6])


```

```python
# creating a league table (one row per team)
# count wins/draws/losses per team
table = pl_1920.groupby('team').agg(
	league=('league', 'first'),
	season=('season', 'first'),
	W=('result', lambda x: (x == 'W').sum()),
	D=('result', lambda x: (x == 'D').sum()),
	L=('result', lambda x: (x == 'L').sum()),
	CS=('ga', 	 lambda x: (x == 0).sum()),
	b6=('b6', 'first'),
	relegated =('is_relegated', 'first')
).reset_index()

#Get points total
table['pts'] = table['W'] * 3 + table['D']

#get cleansheets total


#Give final position ranks based on pts (highest pts -> pos 1), independent of current sort
table['pos'] = table['pts'].rank(method='min', ascending=False).astype(int)

#Sort by points descending for display
table = table.sort_values(by='pts', ascending=False).reset_index(drop=True)

#pts vs big 6
#group games where opponent was in big 6
opp_big6 = pl_1920['opponent'].isin(big_6)

# compute wins and draws vs big 6, then convert to points
b6_table = pl_1920[opp_big6].groupby('team')['result'].agg(
	W_vs_b6=lambda x: (x == 'W').sum(),
	D_vs_b6=lambda x: (x == 'D').sum()
).reset_index()

b6_table['b6_pts'] = b6_table['W_vs_b6'] * 3 + b6_table['D_vs_b6']

# merge b6 points into the main table (keep teams with no b6 matches as 0)
table = table.merge(b6_table[['team', 'b6_pts']], on='team', how='left')
table['b6_pts'] = table['b6_pts'].fillna(0).astype(int)


#Give final position ranks based on pts (highest pts -> pos 1), independent of current sort
table['b6_pos'] = table['b6_pts'].rank(method='min', ascending=False).astype(int)

#pts vs Relegation
# get list of relegated teams (teams marked as relegated in the dataset)
relegated_teams = pl_1920.loc[pl_1920['is_relegated'] == 1, 'team'].unique()

# group games where opponent was a relegated team
opp_rele = pl_1920['opponent'].isin(relegated_teams)

# compute wins and draws vs relegated teams, then convert to points
rele_table = pl_1920[opp_rele].groupby('team')['result'].agg(
	W_vs_rele=lambda x: (x == 'W').sum(),
	D_vs_rele=lambda x: (x == 'D').sum()
).reset_index()

rele_table['rele_pts'] = rele_table['W_vs_rele'] * 3 + rele_table['D_vs_rele']

# merge rele points into the main table (keep teams with no rele matches as 0)
table = table.merge(rele_table[['team', 'rele_pts']], on='team', how='left')
table['rele_pts'] = table['rele_pts'].fillna(0).astype(int)

table['other_pts'] = table['pts'] - (table['rele_pts'] + table['b6_pts'])
table['other_pts'] = table['other_pts'].fillna(0).astype(int)


display(table)
table = table.sort_values('b6_pts', ascending=False)
#Stacked Chart of the teams point sharing
fig = px.bar(
    table,
    x='team',
    y=['other_pts', 'b6_pts', 'rele_pts'],
    title='Points breakdown by opponent group (PL 2019-2020)',
    text_auto=True
    )

fig.update_layout(barmode='stack', xaxis_tickangle=-45)
fig.show()
```

```python
# creating a league table (one row per team) for each season
# count wins/draws/losses per team
table = consolidate.groupby(['team','season']).agg(
	league=('league', 'first'),
	W=('result', lambda x: (x == 'W').sum()),
	D=('result', lambda x: (x == 'D').sum()),
	L=('result', lambda x: (x == 'L').sum()),
	b6=('b6', 'first'),
	relegated =('is_relegated', 'first')
).reset_index()

#Get points total
table['pts'] = table['W'] * 3 + table['D']

#Give final position ranks based on pts for each season (highest pts -> pos 1), independent of current sort
table['pos'] = table.groupby('season')['pts'].rank(method='min', ascending=False).astype(int)


#Sort by points descending for display
table = table.sort_values(by=['pts'], ascending=False).reset_index(drop=True)
table = table.sort_values(by=['season'], ascending=True).reset_index(drop=True)


#pts vs big 6
#group games where opponent was in big 6
opp_big6 = consolidate['opponent'].isin(big_6)

# compute wins and draws vs big 6, then convert to points
b6_table = consolidate[opp_big6].groupby(['team','season'])['result'].agg(
	W_vs_b6=lambda x: (x == 'W').sum(),
	D_vs_b6=lambda x: (x == 'D').sum()
).reset_index()

b6_table['b6_pts'] = b6_table['W_vs_b6'] * 3 + b6_table['D_vs_b6']

# merge b6 points into the main table by team+season (keep teams with no b6 matches as 0)
table = table.merge(b6_table[['team', 'season', 'b6_pts']], on=['team', 'season'], how='left')
table['b6_pts'] = table['b6_pts'].fillna(0).astype(int)


#Give final position ranks based on pts (highest pts -> pos 1), independent of current sort
table['b6_pos'] = table.groupby('season')['b6_pts'].rank(method='min', ascending=False).astype(int)


#pts vs Relegation
# get list of relegated teams per season (teams marked as relegated in each season)
relegated_df = consolidate.loc[consolidate['is_relegated'] == 1, ['season', 'team']].drop_duplicates().rename(columns={'team': 'opponent'})

# mark opponent as relegated in the same season (robustly compute opp_is_relegated)
# build set of season+opponent keys and test membership to avoid KeyError from prior merges/suffixes
_rele_keys = set(relegated_df['season'].astype(str) + "||" + relegated_df['opponent'].astype(str))
_con_keys = consolidate['season'].astype(str) + "||" + consolidate['opponent'].astype(str)
consolidate['opp_is_relegated'] = _con_keys.isin(_rele_keys)

# group games where opponent was a relegated team
opp_rele = consolidate['opp_is_relegated']

# compute wins and draws vs relegated teams, then convert to points
rele_table = consolidate[opp_rele].groupby(['team', 'season'])['result'].agg(
	W_vs_rele=lambda x: (x == 'W').sum(),
	D_vs_rele=lambda x: (x == 'D').sum()
).reset_index()

rele_table['rele_pts'] = rele_table['W_vs_rele'] * 3 + rele_table['D_vs_rele']

# merge rele points into the main table (keep teams with no rele matches as 0)
table = table.merge(rele_table[['team', 'season', 'rele_pts']], on=['team', 'season'], how='left')
table['rele_pts'] = table['rele_pts'].fillna(0).astype(int)

table['other_pts'] = table['pts'] - (table['rele_pts'] + table['b6_pts'])
table['other_pts'] = table['other_pts'].fillna(0).astype(int)

display(table)
#correlation matrix between total points and
```
